```python
import numpy as np
import pandas as pd
import random
from faker import Faker
import string
```

```python
emails = []
amounts = []
accountAs = []
accountBs = []
for i in range(2000): 
    r = random.random() 
    f = Faker()
    An = random.randint(3,12)
    Bn = random.randint(3,12)
    accountA = ''.join(random.choice(string.ascii_uppercase + string.digits) for _ in range(An))
    accountB = ''.join(random.choice(string.ascii_uppercase + string.digits) for _ in range(Bn))
    amount = str(random.randint(2000, 10000000))
    if r < 0.05: 
        name = f.name()
        s = "Hi ,"+name+ " please transfer "+ amount + " from account "+accountA +" to account "+accountB+"."
    elif r >= 0.05 and r < 0.08: 
        s = "Please move $"+amount+" into account "+accountB+ " from "+accountA
    elif r >= 0.08 and r < 0.10: 
        s = "Hello, how are you doing? I need $"+amount + " moved from "+accountA + " to "+accountB+"."
    elif r >= 0.10 and r < 0.12: 
        name = f.name()
        s = name+" I need you to do the following. Please transfer $"+amount+" from "+accountA+ "to "+accountB+". Thank you so much."
    elif r >= 0.12 and r < 0.14: 
        s = "Hi, can you please assist me with the following? I need "+amount+" moved into "+accountB+" from "+ accountA+". Thank you!"
    elif r >= 0.14 and r < 0.16: 
        name = f.name()
        s = "Hi this is "+name+". I need you to do the following. Please help me transfer "+amount +" from "+accountA+" into "+accountB+". Thanks."
    elif r >= 0.16 and r < 0.18: 
        s = "move $"+amount+" from "+accountA+" into account "+accountB+"."
    elif r >= 0.18 and r < 0.20: 
        name = f.name()
        s = "Hey "+ name+", we need to do the following. We need to transfer "+amount+" from account "+accountA+  " into "+accountB+". We need this done ASAP. Thank you."
    elif r >= 0.20 and r < 0.22: 
        s = "We will be transferring $"+amount+"from "+accountA+" to account "+accountB+". Please confirm when this is finished."
    elif r >= 0.22 and r < 0.24: 
        s = "Let's move $"+amount+" into "+accountB+ " from account "+ accountA+". Thanks."
    elif r >= 0.24 and r < 0.26: 
        s = "Transfer "+ amount+" from "+ accountA + " to account "+accountB+"."
    elif r >= 0.26 and r < 0.28: 
        s = "Account "+accountA +" needs $"+amount+" transferred to "+accountB+"."
    elif r >= 0.28 and r < 0.30: 
        s = "We need to remove "+amount+" from "+accountA+" and deposit this into "+accountB+". Thanks :D"
    elif r >= 0.30 and r < 0.34: 
        s = "Hi, can you please assist me with the following? I need $"+amount+" moved into "+accountB+"from "+ accountA+". Thank you very much :D"
    elif r >= 0.34 and r < 0.38: 
        s = "we need to do the following today my friend. Please take $"+amount+" from "+accountA+" and put this in account "+accountB    
    elif r >= 0.38 and r < 0.42: 
        s = "move $"+amount+" to "+accountB+" from "+accountA+". Thanks."
    elif r >= 0.42 and r < 0.46: 
        s = "This needs immediate attention. Please carefully move "+amount+" in the form of US dollars from "+accountA+" into account "+accountB+". Sincerely."
    elif r >= 0.46 and r < 0.50: 
        s = "The following action needs to be taken. Please move "+amount+" from "+accountA+ " to account "+accountB+". Thank you very much."
    elif r >= 0.50 and r < 0.54: 
        name = f.name()
        name2 = f.name()
        s = name+" Please transfer $"+amount+" from "+accountA+ "to "+accountB+". Thank you so much. Best Regards, "+name2
    elif r >= 0.54 and r < 0.58: 
        s = "Let's move "+"$"+amount+" from account "+accountA+" to account "+accountB+". Sound good?"
    elif r >= 0.58 and r < 0.62: 
        s = "Time to move "+amount+" in USD from account "+accountA+" to "+accountB+". Let me know when this is done, thanks!"
    elif r >= 0.62 and r < 0.66: 
        s = "Please do the following: take account "+accountA+" and move $"+amount+" to account "+accountB+"."
    elif r >= 0.66 and r < 0.70: 
        name = f.name()
        name2 = f.name()
        s = "Hey!! "+name+" It's "+name2+". Can you help me do something? I need to transfer $"+amount+" from this account I have called "+accountA+" to my other account titled "+accountB+". Can you do this for me?"
    elif r >= 0.70 and r < 0.74: 
        name = f.name()
        name2 = f.name()
        s = "Hey!! "+name+". Can we do the following? I need to transfer $"+amount+" from this account I have called "+accountA+" to my other account titled "+accountB+". Thanks! Best Regards, "+name2 
    elif r >= 0.74 and r < 0.76: 
        s = "It's that time again! Move "+amount+" from "+ accountA+" to account "+accountB+"Cool?"
    elif r >= 0.76 and r <= 0.80: 
        s = "We need to take the following action immediately! Move $"+amount+" into account "+accountB+ " from account "+ accountA+"!"
    elif r >= 0.80 and r < 0.84: 
        name = f.name()
        s = "Do the following "+name+". Move $"+amount+" from this account called "+accountA+" to account "+accountB+". Thanks. We need this done ASAP. We have a client waiting."
    elif r >= 0.84 and r <0.88: 
        s = 'We will be transferring $'+amount+ "from account "+accountA+ "to account "+accountB+". This needs to be done today. Thanks!"
    elif r >= 0.88 and r< 0.92: 
        s = "Today we need to move $"+amount+" from account "+accountA+ " to this account titled "+accountB+".Thanks so much."   
    elif r >= 0.92 and r < 0.96: 
        name = f.name()
        s = "Hey "+name+"! Please move $"+amount+" from account "+accountA + " to account "+accountB+". Thanks! Send me an email back when it's done! Thanks :D :D"
    else: 
        s = "Move money from account "+accountA + " to account "+accountB+ " of the amount $"+amount+". Thank you!"
    
    emails.append(s)
    amounts.append(amount)
    accountAs.append(accountA)
    accountBs.append(accountB)
#df = pd.DataFrame()
#df['email'] = emails
#df['amount'] = amounts
#df['accountA'] = accountAs
#df['accountB'] = accountBs
#df.to_csv('data.csv',sep="|")
```

```python
dta = list(zip(emails,amounts,accountAs,accountBs))
df = pd.DataFrame(dta)
df = df.rename(columns={0:"email",1:"amount",2:"accountA",3:"accountB"})
df.to_csv('data.csv',sep="|",index=None)
```
